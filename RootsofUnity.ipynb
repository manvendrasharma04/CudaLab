```python
import numpy as np
import matplotlib.pyplot as plt
from math import gcd

def roots_of_unity(n):
    """
    Compute all n-th roots of unity.
    """
    return [np.exp(2j * np.pi * k / n) for k in range(n)]

def primitive_roots_of_unity(n):
    """
    Compute n-th primitive roots of unity.
    """
    all_roots = roots_of_unity(n)
    return [all_roots[k] for k in range(n) if gcd(k, n) == 1]

def plot_roots_and_primitive_roots(n):
    """
    Plot n-th roots of unity and n-th primitive roots of unity on the complex plane.
    Use different colors to distinguish them.
    """
    all_roots = roots_of_unity(n)
    primitive_roots = primitive_roots_of_unity(n)
    
    plt.figure(figsize=(8, 8))
    plt.axhline(0, color="gray", linewidth=0.5)
    plt.axvline(0, color="gray", linewidth=0.5)
    plt.grid(color="lightgray", linestyle="--", linewidth=0.5)
    
    # Unit circle
    circle = plt.Circle((0, 0), 1, color="blue", fill=False, linestyle="--", linewidth=0.8)
    plt.gca().add_artist(circle)
    
    # Plot all roots of unity
    for root in all_roots:
        plt.plot(root.real, root.imag, 'o', color="orange", label="All roots" if all_roots.index(root) == 0 else None)
    
    # Plot primitive roots of unity
    for root in primitive_roots:
        plt.plot(root.real, root.imag, 'o', color="green", label="Primitive roots" if primitive_roots.index(root) == 0 else None)
    
    # Title and legend
    plt.title(f"{n}-th Roots of Unity and Primitive Roots of Unity")
    plt.xlabel("Re")
    plt.ylabel("Im")
    plt.axis("equal")
    plt.legend()
    plt.show()

# Example usage for n=7
plot_roots_and_primitive_roots(34)

```
